```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("tempno.db")

missions_df = pd.read_sql("SELECT * FROM missions", conn)
dossiers_df = pd.read_sql("SELECT * FROM dossiers", conn)
clients_df = pd.read_sql("SELECT * FROM clients", conn)
```

```python
df = missions_df.merge(dossiers_df, left_on="id_dossier",
                                right_on="id", suffixes=("", "_dossier")
                               )
```

```python
df
```

```python
df = df.merge(clients_df, left_on="id_client",
                                right_on="id", suffixes=("", "_client")
                               )
```
