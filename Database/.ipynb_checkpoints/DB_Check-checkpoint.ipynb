Visualiser la base

```python
import psycopg2
from psycopg2 import sql
import streamlit as st
import pandas as pd

# Connexion centralis√©e et mise en cache pour √©viter les recharges inutiles
@st.cache_resource
def get_connection():
    return psycopg2.connect(
        dbname="tempno",
        user="postgres",
        password="Bef3orlaihec",
        host="localhost",
        port="5432"
    )

# Visualisation d'une table
def viz(nom_table):
    conn = get_connection()
    try:
        query = sql.SQL("SELECT * FROM {}").format(sql.Identifier(nom_table))
        df = pd.read_sql_query(query.as_string(conn), conn)
        return df
    except Exception as e:
        st.error(f"Erreur lors de la lecture de la table '{nom_table}' : {e}")
        return pd.DataFrame()
    finally:
        conn.close()

#Affichage des tables
def view_tables():
    conn = get_connection()
    try:
        cursor = conn.cursor()
        cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema='public'")
        tables = cursor.fetchall()
        for table in tables:
            st.write("üìÑ Table :", table[0])
    finally:
        conn.close()

# Supprime les donn√©es d'un utilisateur dans une table
def clear_table(nom_table):
    conn = get_connection()
    try:
        cursor = conn.cursor()
        query = sql.SQL("DELETE FROM {}").format(sql.Identifier(nom_table))
        cursor.execute(query)
        conn.commit()
        df = pd.read_sql_query(
            sql.SQL("SELECT * FROM {}").format(sql.Identifier(nom_table)).as_string(conn), conn
        )
        st.success(f"Donn√©es supprim√©es dans '{nom_table}'.")
        st.dataframe(df)
    except Exception as e:
        st.error(f"Erreur lors de la suppression : {e}")
    finally:
        conn.close()

# Supprime compl√®tement une table
def drop_table(nom_table):
    conn = get_connection()
    try:
        cursor = conn.cursor()
        query = sql.SQL("DROP TABLE {}").format(sql.Identifier(nom_table))
        cursor.execute(query)
        conn.commit()
        st.success(f"Table '{nom_table}' supprim√©e.")
    except Exception as e:
        st.error(f"Erreur lors de la suppression de la table : {e}")
    finally:
        conn.close()
```

```python
conn = get_connection()
cursor = conn.cursor()

query = sql.SQL("ALTER TABLE {} DROP COLUMN {}").format(
    sql.Identifier("dossiers"),
    sql.Identifier("code_client")
)

cursor.execute(query)
conn.commit()
cursor.close()
conn.close()
```
