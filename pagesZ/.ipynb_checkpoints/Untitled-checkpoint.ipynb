```python
import psycopg2
import datetime
import pandas as pd
from psycopg2 import sql
from datetime import date

def get_connection():
    return psycopg2.connect(
        dbname=os.getenv("DB_NAME"),
        user=os.getenv("DB_USER"),
        password=os.getenv("DB_PASSWORD"),
        host=os.getenv("DB_HOST"),
        port=os.getenv("DB_PORT")
    )

def lire_saisies_mois(utilisateur_id):
    today = date.today()
    first_day = today.replace(day=1)
    last_day = today.replace(day=31) if today.month == 12 else \
               today.replace(day=1, month=today.month + 1) - pd.Timedelta(days=1)

    with get_connection() as conn:
        query = """
        SELECT hs.id, hs.date, hs.type, hs.duree_heures, m.matricule_mission, c.nom_client, hs.code, hs.commentaire
        FROM heures_saisies AS hs
        LEFT JOIN missions m ON hs.mission_id = m.id
        LEFT JOIN clients c ON hs.client_id = c.id
        WHERE hs.utilisateur_id = %s AND hs.date BETWEEN %s AND %s
        """
        df = pd.read_sql(query, conn, params=(utilisateur_id, first_day, last_day))
    return df

df = lire_saisies_mois(1)
```

```python
def odd(number):
    input_num = number

    if input_num % 2 == 0:
        print(f"{input_num} est un nombre pair")
    else:
        print(f"{input_num} est un nombre impair")
```

```python
def square():
    number=int(input("Entrez un nombre"))
    list = []
    for i in range(number+1):
        square = i**2
        list.append(square)
    print(f"Les carrés des nombres de 0 à {number} sont :")
    return(list)
```

```python
def sum_even(list_num):
    even = []
    odd = []
    for i in (list_num):
        if i%2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return [sum(i for i in even)]
```

```python
def anag(mot1, mot2):
    mot1 = str.lower(mot1)
    mot2 = str.lower(mot2)

    sort_1 = sorted(mot1)
    sort_2 = sorted(mot2)

    if sort_1 == sort_2:
        print(f"{mot1} et {mot2} sont des anagrammes")
    else:
        print(f"{mot1} et {mot2} ne sont pas des anagrammes")
```

```python
anag("chien", "ananas")
```

```python
def calc_cli():
    print("Sélectionnez un premier nombre :")
    inp1 = int(input())
    print("Sélectionnez un second nombre :")
    inp2 = int(input())
    print("Sélectionnez une opération (+, -, *, /) :")
    op = input()

    if op == "+":
        result = inp1 + inp2
    elif op == "-":
        result = inp1 - inp2
    elif op == "*":
        result = inp1 * inp2
    elif op == "/" and inp2 != 0:
        result = inp1 / inp2
    else:
        result = "Opération non reconnue, veuillez réessayer"
    return result
```

```python
calc_cli()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

phrase_2 = "Mon chien s'appelle Nanuuk. Nanuuk est un malamute d'Alaska"

def most_frequent(phrase):
    stop_words = set(stopwords.words('french'))

    phrase = phrase.lower().replace("'", " ")
    phrase = re.sub(r"[^\w\s]", "", phrase)
    phrase = phrase.split()

    phrase = [mot for mot in phrase if mot not in stop_words]

    count = Counter(phrase)

    most_freq = max(count.items(), key=lambda x : x[1])

    return most_freq, phrase
```

```python
most_frequent(phrase_2)
```

```python
def juste_prix():
    prix = np.random.randint(1, 101)
    guess = int(input("Veuillez entrer un nombre compris entre 1 et 100 :" ))
    while guess != prix:
        if guess > prix:
            print("Moins")
            guess = int(input("Veuillez entrer un nombre compris entre 1 et 100"))
        elif guess < prix:
            print("Plus")
            guess = int(input("Veuillez entrer un nombre compris entre 1 et 100"))

    if guess == prix:
        print(f"Félicitations ! Le prix exact était de {prix}")
```
