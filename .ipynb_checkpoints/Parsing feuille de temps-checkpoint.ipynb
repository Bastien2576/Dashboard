```python
import pandas as pd
import numpy as np
import time
import warnings

warnings.filterwarnings('ignore')

path = "C:/Users/ubamiot/Scripts/Application_GI/Database/"

xls = pd.ExcelFile(path + "Fichier test.xlsx")

dfs = {sheetname: pd.read_excel(xls, sheet_name = sheetname, header=4)
      for sheetname in xls.sheet_names
      if sheetname.startswith("Semaine")}
```

```python
def wide_to_long(df: pd.DataFrame, fixed_cols_n: int = 6) -> pd.DataFrame:
    fixed_cols = list(df.columns[:fixed_cols_n])
    day_cols = list(df.columns[fixed_cols_n:])

    df2 = df.copy()
    df2.columns = [str(c) for c in df2.columns]

    fixed_cols = [str(c) for c in fixed_cols]
    day_cols = [str(c) for c in day_cols]

    long_df = df2.melt(
        id_vars=fixed_cols,
        value_vars=day_cols,
        var_name="Date",
        value_name="Heures",
    )

    long_df["Heures"] = pd.to_numeric(long_df["Heures"], errors="coerce")

    long_df = long_df.drop(columns=['Total du mois'])
    long_df = long_df.dropna(subset=["Heures"])
    long_df = long_df[long_df["Heures"] != 0]

    long_df = long_df[['Date', 'N° Client', 'Nom du client', 'Mission', 'Tâche', 'Détail', 'Heures']]

    return long_df.reset_index(drop=True)
```

```python
fact_l = []
nfact_l = []

for name, df in dfs.items():

    dates = df.columns[6:13]
    df.rename(
        columns={
            c: pd.to_datetime(c, errors='coerce')
            for c in dates
        },
        inplace=True
    )

    df.rename(
        columns={
            c: c.strftime(format='%Y-%m-%d') for c in df.columns[6:13]
            if isinstance(c, pd.Timestamp)
        },
        inplace=True
    )

    df = df.rename(columns=({
            'du': 'N° Client',
            'du.1': 'Total du mois'
    }))

    fact = df.iloc[1:40,:13]
    fact.iloc[:,6:13] = fact.iloc[:,6:13].astype(float)
    fact = fact.dropna(subset="N° Client")

    try:
        fact = wide_to_long(fact, 6)
        fact['Date'] = pd.to_datetime(fact['Date'], dayfirst=False)
        fact_l.append(fact)
    except:
        print(f"Pas d'heures facturables répertoriées pour la {name}")

    nfact = df.iloc[41:54,:13]
    nfact = nfact.loc[nfact['Total du mois'] != 0]

    nfact = nfact.drop(
        columns=[c for c in nfact.columns[6:13] if nfact[c].isna().all()]
    ).fillna(0)
    nfact.iloc[:,6:13] = nfact.iloc[:,6:13].astype(float)

    try:
        nfact = wide_to_long(nfact, 6)
        nfact['Date'] = pd.to_datetime(nfact['Date'], dayfirst=False)
        nfact_l.append(nfact)
    except:
        print(f"Pas d'heures non facturables répertoriées pour la {name}")
```
